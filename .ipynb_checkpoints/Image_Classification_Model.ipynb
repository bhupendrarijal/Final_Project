Importing Libraries

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import Rescaling
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Dense, Activation, BatchNormalization, GlobalAveragePooling2D
```

Importing Datasets

```python
data_train_path = 'Fruits_Vegetables_large/train'
data_test_path = 'Fruits_Vegetables_large/test'
data_validation_path = 'Fruits_Vegetables_large/validation'
```

**Exploratory Data Analysis (EDA)**

```python
train_path = "Fruits_Vegetables_large/train"
class_counts = {}

for class_name in os.listdir(train_path):
    class_folder = os.path.join(train_path, class_name)
    if os.path.isdir(class_folder):
        count = len([f for f in os.listdir(class_folder) if os.path.isfile(os.path.join(class_folder, f))])
        class_counts[class_name] = count

for class_name, count in class_counts.items():
    print(f"{class_name}: {count} images")
```

Image Data Preprocessing

```python
img_width = 180
img_height = 180
batch_size = 32
img_channels=3
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
```

Count of Classes and Images on Training Dataset

```python
data_train = tf.keras.utils.image_dataset_from_directory(
    data_train_path,
    shuffle=True,
    image_size = (img_width, img_height),
    validation_split = False)
```

Count of Classes and Images on Validation Dataset

```python
data_val = tf.keras.utils.image_dataset_from_directory(
    data_validation_path,
    shuffle = True,
    image_size = (img_width, img_height),
    validation_split = False)
```

Count of Classes and Images on Test Dataset

```python
data_test = tf.keras.utils.image_dataset_from_directory(
    data_test_path,
    shuffle = True,
    image_size = (img_width, img_height),
    validation_split = False)
```

```python
data_train
```

Classes in the Dataset

```python
data_category = data_train.class_names
print("Categories:", data_category)
```

Sample Images on Training Datasets

```python
plt.figure(figsize=(12, 5))
for images, labels in data_train.take(1):
    for i in range(18):
        plt.subplot(3, 6, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(data_category[labels[i]])
        plt.axis("off")
plt.tight_layout()
plt.show()
```

Sample Images on Test Datasets

```python
plt.figure(figsize=(12, 5))
for images, labels in data_test.take(1):
    for i in range(12):
        plt.subplot(2, 6, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(data_category[labels[i]])
        plt.axis("off")
plt.tight_layout()
plt.show()
```

Sample Images on Validation Datasets

```python
plt.figure(figsize=(12, 5))
for images, labels in data_val.take(1):
    for i in range(12):
        plt.subplot(2, 6, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(data_category[labels[i]])
        plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
AUTOTUNE = tf.data.AUTOTUNE
data_train = data_train.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
data_val = data_val.cache().prefetch(buffer_size=AUTOTUNE)
data_test = data_test.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
model = Sequential()

# Normalize input
model.add(Input(shape=(img_width, img_height, img_channels)))

# First Set of Convolution and Pooling Layer
model.add(Conv2D(32, kernel_size=(3,3), activation='relu',padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))

# Second Set of Convolution and Pooling Layer
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu',padding='same' ))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))

# Third Set of Convolution and Pooling Layer
model.add(Conv2D(128, kernel_size=(3, 3), activation='relu',padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))

# Fully Connected Layer
model.add(GlobalAveragePooling2D())
model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(units = len(data_category), activation='softmax'))
```

```python
model.summary()
```

Compiling and Training Phase

```python
model.compile(optimizer= 'adam',
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
early_stop = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    mode='min',
    restore_best_weights=True, 
    verbose=1,
    patience=5)

training_history = model.fit(x = data_train, 
                             validation_data = data_val, 
                             epochs = 25, 
                             callbacks=[early_stop])
```

Saving Model

```python
model.save('Image_classification_model.keras')
```

Recording History

```python
training_history.history
```

```python
import json
with open ('training_hist.json','w') as f:
    json.dump(training_history.history,f)
```

Calculating accuracy of Model Achieved on validation set

```python
print('validation set Accuracy: {} %'.format(training_history.history['val_accuracy'][-1]*100))
```

Accuracy Visualization

```python
plt.figure(figsize=(12, 5))

# Accuracy
plt.subplot(1, 2, 1)
plt.plot(training_history.history['accuracy'], label='Training Accuracy')
plt.plot(training_history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

# Loss
plt.subplot(1, 2, 2)
plt.plot(training_history.history['loss'], label='Training Loss')
plt.plot(training_history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
test_loss, test_accuracy = model.evaluate(data_test)
print(f'Test Accuracy: {test_accuracy * 100:.2f}%')
```

Testing Model

```python
predict = model.predict(img_bat)
predicted_class = np.argmax(predict)
confidence = np.max(predict) * 100
```

```python
print('Veg/Fruit in Image is "{}" with accuracy of {:.2f}%'.format(data_category[predicted_class], confidence))
```
